```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Droping 'PassengerId', 'Name', 'Ticket', and 'Embarked' columns.
df.drop(['PassengerId', 'Name', 'Ticket', 'Embarked','Cabin'], axis=1, inplace=True)
```

```python
df.describe()
```

```python
# Convert the 'Sex' column to numerical values
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df["Sex"].unique
```

```python
df.head(6)
```

```python
# Create a bar plot
sns.set(style="whitegrid")  # Set the style of the plot
plt.figure(figsize=(8, 6))  # Set the figure size
sns.barplot(x="Sex", y="Survived", data=df)
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender")

# Show the plot
plt.show()
```

The Titanic dataset, sourced from Kaggle, records key information about passengers aboard the Titanic, including demographics, ticket class, fare paid, and survival status. This report aims to provide an initial analysis of the dataset to uncover insights into factors influencing survival rates among passengers.

## Observations
Upon initial exploration of the Titanic dataset, several key insights have emerged:
Survival Rates by Gender:
Females had a significantly higher survival rate compared to males. Approximately 74% of females survived, whereas only around 19% of males survived.
Survival Rates by Passenger Class:
Passengers in first class had the highest survival rate among all classes, followed by second class, and then third class. This suggests a possible correlation between socio-economic status and survival likelihood.

## Age Distribution:
The age distribution of passengers shows a predominant presence of young adults, with the average age being approximately 29 years. There is a notable presence of children and elderly individuals as well.

## Conclusion
In summary, this initial review of the Titanic dataset highlights significant factors influencing survival rates, namely gender and passenger class. Females and passengers in higher classes had higher chances of survival.
